# Numerical integration

```python
from sympy.printing.pretty.pretty_symbology import xstr
from array import *
import math
import matplotlib.pyplot as plt
from sympy import *
from scipy import integrate
```

```python
def f(x):
  return sqrt(2-x**2)

(v, err) = integrate.quad(f, 0, 1)

def fun(x):
  return sqrt(2-x**2)
```

```python
n = 1000
left_limit = 0
right_limit = 1
lenth = right_limit - left_limit 
h = lenth/(n)
print("шаг = ", h)
integral = 0
x = []
```

## Riemann

```python
for i in range(0, n+1):
    x.append(left_limit + (lenth*i)/n)

for i in range(0, n):
    integral += fun((x[i+1]+x[i])/2)*h

print("интеграл равен = ", integral)
print("ошибка = ", abs(integral-v)) 
```

```python
mistake = []
xMistake = []

for n1 in range(5,100):
  h1 = lenth/(n1)
  integral1 = 0
  x1 = []

  for i in range(0, n1+1):
    x1.append(left_limit + (lenth*i)/n1)

  for i in range(0, n1):
    integral1 += fun((x1[i+1]+x1[i])/2)*h1

  mistake.append(abs(integral1-v))
  xMistake.append(n1)


plt.plot(xMistake, mistake)
plt.show()
```

## Trapezoid's Rule

```python

n2 = n
integral2 = 0
x2 = []

for i in range(0, n2+1):
    x2.append(left_limit + (lenth*i)/n2)

for i in range(0, n2):
  integral2 += (fun(x2[i])+fun(x2[i+1]))*(x2[i+1]-x2[i])/2

print("интеграл равен = ", integral2)
print("ошибка = ", abs(integral2-v)) 
```

```python
mistake2 = []
xMistake2 = []

for n2 in range(5,100):
  h2 = lenth/(n2)
  integral2 = 0
  x2 = []

  for i in range(0, n2+1):
      x2.append(left_limit + (lenth*i)/n2)

  for i in range(0, n2):
    integral2 += (fun(x2[i])+fun(x2[i+1]))*(x2[i+1]-x2[i])/2

  mistake2.append(abs(integral2-v))
  xMistake2.append(n2)


plt.plot(xMistake2, mistake2)
plt.show()
```

## Simpson's rule

```python
n3 = 1000
h3 = h
integral3 = 0
x3 = []

for i in range(0, n3+1):
  x3.append(left_limit + (lenth*i)/n3)

for i in range(0, n3):
  integral3 += (h3/6)*(fun(x3[i+1])+4*fun((x3[i+1]+x3[i])/2)+fun(x3[i]))

print("интеграл равен = ", integral3)
print("ошибка = ", abs(integral3-v)) 
```

```python
mistake3 = []
xMistake3 = []

for n3 in range(5,100):
  h3 = lenth/(n3)
  integral3 = 0
  x3 = []

  for i in range(0, n3+1):
    x3.append(left_limit + (lenth*i)/n3)

  for i in range(0, n3):
    integral3 += (h3/6)*(fun(x3[i+1])+4*fun((x3[i+1]+x3[i])/2)+fun(x3[i]))

  mistake3.append(abs(integral3-v))
  xMistake3.append(n3)


plt.plot(xMistake3, mistake3)
plt.show()
```
